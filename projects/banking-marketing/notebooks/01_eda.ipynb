4.y(マーケティングの成否)とそれぞれの特徴量の関係

```python
# age
df = df_eda.copy()

# 5歳刻みで見てみる
df["age"] = (df["age"]//5)*5

sr_result = df.groupby("age")["y"].value_counts()

df_result = sr_result.to_frame()
df_result = df_result.reset_index()


for col in df_result["y"].unique():
    df_plot = df_result[df_result["y"]==col]

    labels = df_plot["age"]
    x = df_plot["count"]

    plt.pie(x, labels=labels, autopct="%1.1f%%", startangle=90)
    plt.title(col)
    plt.show()

# マーケティングの成否が特定の年齢層に偏ってはいなさそう
```

```python
# job
df = df_eda.copy()

sr_result = df.groupby("job")["y"].value_counts()

df_result = sr_result.to_frame()
df_result = df_result.reset_index()


for col in df_result["y"].unique():
    df_plot = df_result[df_result["y"]==col]

    labels = df_plot["job"]
    x = df_plot["count"]

    plt.pie(x, labels=labels, autopct="%1.1f%%", startangle=90)
    plt.title(col)
    plt.show()

# 全体の構成の割合は大きく変わらなそう
```

```python
# グラフだけではなく、数値で見てみる
# 各jobごとのyes or noを見る

df = df_eda.copy()

sr_result = df.groupby("job")["y"].value_counts(normalize=True)

df_result = sr_result.to_frame()
df_result = df_result.reset_index()

df_difference = pd.DataFrame(columns=["job", "diff"])

list_data= []

for col in df_result["job"].unique():

    yes_proportion = df_result[(df_result["job"]==col) & (df_result["y"]=="yes")]["proportion"].values[0]
    no_proportion = df_result[(df_result["job"]==col) & (df_result["y"]=="no")]["proportion"].values[0]

    list_data.append([col, yes_proportion, no_proportion])

df_difference = pd.DataFrame(data=list_data, columns=["job", "yes", "no"])

df_difference["diff"] = df_difference["no"] - df_difference["yes"]
```

```python
# no-yesの差が小さいjob順

df_difference.sort_values("diff")

# 差が小さいほど、該当のjobはyがyesになりやすい可能性(差を求めずに、普通にyesの割合だけでいい気もする)
# 全体のyes or noの割合は、yes:no = 11.7%:88.3%
```

```python
# marital
df = df_eda.copy()

sr_result = df.groupby("marital")["y"].value_counts()

df_result = sr_result.to_frame()
df_result = df_result.reset_index()


for col in df_result["y"].unique():
    df_plot = df_result[df_result["y"]==col]

    labels = df_plot["marital"]
    x = df_plot["count"]

    plt.pie(x, labels=labels, autopct="%1.1f%%", startangle=90)
    plt.title(col)
    plt.show()

# marriedの方がnoの割合が9.2%高く、singleはyesの割合が9%高い
# divocedは差がない
# singleの方が、マーケティングが成功する確率が高い、可能性あり
```

```python
# education

df = df_eda.copy()

sr_result = df.groupby("education")["y"].value_counts()

df_result = sr_result.to_frame()
df_result = df_result.reset_index()


for col in df_result["y"].unique():
    df_plot = df_result[df_result["y"]==col]

    labels = df_plot["education"]
    x = df_plot["count"]

    plt.pie(x, labels=labels, autopct="%1.1f%%", startangle=90)
    plt.title(col)
    plt.show()

# 学歴は割と差がありそう
```

```python
# balance
# 基礎統計量再掲
# count     45211.000000
# mean       1362.272058
# std        3044.765829
# min       -8019.000000
# 25%          72.000000
# 50%         448.000000
# 75%        1428.000000
# max      102127.000000

df = df_eda.copy()

# 全体のバランスを見たいので、
# 一旦上限を90%のところでまるめこむ
ceiling_value = df["balance"].quantile(0.9)
df.loc[df["balance"] >= ceiling_value, "balance"] = ceiling_value

# 一旦下限を10%のところでまるめこむ
floar_value = df["balance"].quantile(0.1)
df.loc[df["balance"] <= floar_value, "balance"] = floar_value

# 一旦250刻みにする
df["balance"] = (df["balance"]//250)*250

sr_result = df.groupby("balance")["y"].value_counts()

df_result = sr_result.to_frame()
df_result = df_result.reset_index()


for col in df_result["y"].unique():
    df_plot = df_result[df_result["y"]==col]

    labels = df_plot["balance"]
    x = df_plot["count"]

    plt.pie(x, labels=labels, autopct="%1.1f%%", startangle=90)
    plt.title(col)
    plt.show()

# balanceが0~250の間のnoが、yesに比べて12.6%高い
```
